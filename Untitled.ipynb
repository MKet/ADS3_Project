```python
#ext = interpolate.interp1d(years, plotdata, fill_value = "extrapolate")
def extrapolatePlot(data, years):
    plt.figure(1, figsize=(12,12))
    plotname = data.name
    plt.suptitle("Predictions for " +plotname)
    plt.bar(range(years[0], years[len(years)-1]+1), data, label="Actual Data", color=colors[0])
    z = np.polyfit(range(years[0], years[len(years)-1]+1), data, 1)
    ext = np.poly1d(z)
    prevdata = data.iloc[-1]
    extrapolatedlin = []
    extrayears = []
    for i in range(0, len(list(range(years[0],2019))) - len(data)):
        extrapolatedlin.append(ext(2018 - i))
        extrayears.append(2018 -i - 0.25)
    plt.figure(1)
    plt.bar(extrayears, extrapolatedlin, label="Predicted Linear", color=colors[len(colors)-1], width=0.30, alpha=0.4)
    plt.plot(extrayears, extrapolatedlin, label="Predicted Linear", color=colors[len(colors)-1])
    trendline = []
    for year in years:
        trendline.append(ext(year))
    trendlineyears = years.copy()
    trendlineyears.append(extrayears[len(extrayears)-1])
    trendline.append(extrapolatedlin[len(extrapolatedlin)-1])
    plt.plot(trendlineyears, trendline, linestyle="-", color=colors[len(colors)-1], alpha=0.4,dashes=[10,10])
    z = np.polyfit(range(years[0], years[len(years)-1]+1), data, 2)
    ext = np.poly1d(z)
    prevdata = data.iloc[-1]
    extrapolatedlin = []
    extrayears = []
    for i in range(0, len(list(range(years[0],2019))) - len(data)):
        extrapolatedlin.append(ext(2018 - i))
        extrayears.append(2018 -i + 0.25)
    plt.figure(1)
    plt.bar(extrayears, extrapolatedlin, label="Predicted Quadratic", color=colors[len(colors)-4], width=0.30, alpha=0.4)
    plt.plot(extrayears, extrapolatedlin, label="Predicted Quadratic", color=colors[len(colors)-4])
    trendline = []
    for year in years:
        trendline.append(ext(year))
    trendline.append(extrapolatedlin[len(extrapolatedlin)-1])
    trendlineyears = years.copy()
    trendlineyears.append(extrayears[len(extrayears)-1])
    plt.plot(trendlineyears, trendline, linestyle='-', color=colors[len(colors)-4], alpha=0.4, dashes=[10,10])
    plt.figure(1)
    plt.xlabel("Years")
    plt.ylabel(plotname)
    plt.legend()
    
    
    
yearlist = list(range(2005, 2016))
# print(plotdata[1:])
# print(yearlist[1:])
extrapolatePlot(plotdata[:len(yearlist)-1], yearlist[:len(yearlist)-1])
#Beide modellen mogen gebruikt worden als hier de juiste uitleg bij gegeven is.
#Uitvinden welk getal linear en kwadratisch is en deze beide laten zien in een vakje.
#Met subplots beide predictions over elkaar zetten.
#VEROORLOVING GEVEN IS OOK GOED ALS ALLES FOUT GAAT.
#Graphs uitleggen in het rapport.
```

```python
def averagePerYear(df, column):
    averages = []
    years = list(range(2005, 2016))
    for i in range(0, len(years)):
        averages.append([])
        #For loop door agency ids
        for agency in df["AGENCY_ID"].unique():
            df_agency = df[(df["AGENCY_ID"] == agency) & (df["STAT_PROFILE_DATE_YEAR"] == years[i])]
            total = 0
            #For loop door producten
            for productid in df_agency["PROD_ID"].unique():
                total = total + df_agency[(df_agency["PROD_ID"] == productid)]["POLY_INFORCE_QTY"]
            averages[i].append(total)
    return averages
print(averagePerYear(df, ""))
```
