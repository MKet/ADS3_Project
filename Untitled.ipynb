```python
import sklearn as sk
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from scipy import interpolate

print('numpy version:', np.__version__)
print('pandas version:', pd.__version__)
print('scikit-learn version:', sk.__version__)
print('matplotlib version:', matplotlib.__version__)

%matplotlib inline
```

```python
df = pd.read_csv("cleaned_PL_dataset_informa.csv", low_memory=False, encoding = 'ISO-8859-1')
```

```python
df.head(500)
```

```python
testdata = df[(df["AGENCY_ID"] == 32) & (df["PROD_ABBR"] == "MOTORHOME")& (df["STATE_ABBR"] == "OH") & (df["STAT_PROFILE_DATE_YEAR"] < 2011)]
testdata.head(50)
```

```python
plotdata = testdata["WRTN_PREM_AMT"]
years = testdata["STAT_PROFILE_DATE_YEAR"]

plt.plot(years, plotdata)
plt.show()
```

```python
ext = interpolate.interp1d(years, plotdata, fill_value = "extrapolate")

z = np.polyfit(years, plotdata, 1)
ext = np.poly1d(z)
yearsextra = years.copy()
extrapolated = plotdata.copy()
extradata = [ext(2012), ext(2014), ext(2016), ext(2017), ext(2018)]
for index in range(len(extradata)):
    if (extradata[index] < 0):
        extradata[index] = 0
extrapolated =extrapolated.append(pd.Series(extradata), ignore_index=True)
yearsextra = yearsextra.append(pd.Series([2012, 2014, 2016, 2017, 2018]), ignore_index=True)
plt.plot(yearsextra, extrapolated)
plt.show()
```
